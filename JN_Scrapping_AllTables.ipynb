```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import datetime
```

```python
def extract_tables(url):
  # Send a GET request to the URL
  response = requests.get(url)

  # Parse the HTML content of the page
  soup = BeautifulSoup(response.content, 'html.parser')

  # Find all table elements on the page
  tables = soup.find_all('table')

  # Initialize an empty list to store the dataframes
  dataframes = []

  # Iterate over the tables
  for table in tables:
    # Convert the table to a dataframe
    df = pd.read_html(str(table))[0]

    # Add the dataframe to the list
    dataframes.append(df)

  # Return the list of dataframes
  return dataframes
```

```python
def convert_price_string_to_float(df, column_name):
    # extract the numerical part of the string
    df[column_name] = df[column_name].str.extract('(\d+.\d+)')
    # replace the ',' with '.'
    df[column_name] = df[column_name].str.replace(',', '.')
    # convert the column to float
    df[column_name] = df[column_name].astype(float)
    return df
def convert_price_string_to_float_v2(df, column_index):
    # extract the numerical part of the string
    df.iloc[:, column_index] = df.iloc[:, column_index].str.extract('(\d+.\d+)')
    # replace the ',' with '.'
    df.iloc[:, column_index] = df.iloc[:, column_index].str.replace(',', '.')
    # convert the column to float
    df.iloc[:, column_index] = df.iloc[:, column_index].astype(float)
    return df
```

```python
def extract_classes(url, class_name):
    # Get the HTML content of the page
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    # Extract all elements with the given class name
    elements = soup.find_all(class_=class_name)
    return elements
def extract_job_data_nestle():
    url = 'https://www.nestle.com/jobs/search-jobs?keyword=marketing&country=CH&location=&career_area=4351&sort_bef_combine=created_DESC'
    jc = extract_classes(url, 'jobs-card')
    job_title_list = []
    job_link_list = []
    location_list = []
    for job in jc:
        job_title = job.find('div', class_='jobs-title')
        job_title1 = job_title.find('a').text
        job_link = job_title.a.attrs['href']
        location = job.find('div', class_='jobs-location').get_text(strip=True)
        job_title_list.append(job_title)
        job_link_list.append(job_link)
        location_list.append(location)
    data = {'job_title': job_title_list, 'job_link': job_link_list, 'location': location_list}
    df = pd.DataFrame(data)
    return df
```

```python
import smtplib

def send_email(to_email, subject, body):
    FROM = 'your_email@example.com'
    TO = to_email
    SUBJECT = subject
    BODY = body
    PASSWORD = 'your_email_password'

    # Prepare the message
    message = f"""\
    From: {FROM}
    To: {TO}
    Subject: {SUBJECT}
    {BODY}"""

    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.ehlo()
        server.starttls()
        server.login(FROM, PASSWORD)
        server.sendmail(FROM, TO, message)
        server.close()
        print(f'Email sent to {TO}!')
    except Exception as e:
        print(f'Something went wrong... {e}')
def new_elements(df1, df2):
    new_elements = pd.concat([df2, df1]).drop_duplicates(keep=False)
    return new_elements
def send_email_with_new_data(df1, df2, to_email):
    new_data = new_elements(df1, df2)
    if new_data.empty:
        print("No new data")
    else:
        subject = "New Job Offer"
        body = "New Job Offers:\n" + new_data.to_string()
        send_email(to_email, subject, body)
```

```python
def get_file_creation_date(file_path):
    if os.path.exists(file_path):
        ctime = os.path.getctime(file_path)
        return datetime.fromtimestamp(ctime)
    else:
        raise FileNotFoundError(f"{file_path} does not exist.")
```
